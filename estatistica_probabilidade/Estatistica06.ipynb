```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv") #csv do governo americano sobre eficiencia dos combustiveis

gear_counts = df['# Gears'].value_counts()#o pandas vai contar um valor especifico que no caso e o numero de marchas mais comuns

gear_counts.plot(kind='bar') #tracamos um grafico do estilo barras
```

```python
#Uma outra biblioteca para tracar os graficos e a Seaborn, que é simplesmente o matplot mais moderno
import seaborn as sns
sns.set
gear_counts.plot(kind='bar')
```
