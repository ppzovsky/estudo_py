```python
#A densidade da propabilidade, é a chance de cada faixa de valores saírem em uma distribuição
#se estamos trabalhando com dados discretos (idade por exemplo), essa função é simplificada e se chama função de massa de propabilidade
```

```python
#Um site recebe uma media de 500 visitas por dia. Qual a probabilidade dele receber 550 visitas em um dia?
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

media_acesso = 500 #numero medio de pessoas que acessam o site por dia
x = np.arange(400, 600, 0.5) #criar um vetor para anotar todos os numeros entre 400 e 600
plt.plot(x, poisson.pmf(x, media_acesso)) #plotamos um grafico com a distribuicao de poisson que vai desenhar as probabilidades da amosta
```

```python
#Com o grafico plotado, podemos perceber a propabilidade de 550, q e +- 0.002, ou seja 2%
```
