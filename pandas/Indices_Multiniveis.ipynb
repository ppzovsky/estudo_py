```python
import numpy as np
import pandas as pd
```

```python
#Niveis de indice

outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
```

```python
outside
```

```python
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
#cria um dataframe de multiniveis
#serve para categorizar os dataframes de maneira melhor
```

```python
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6,2), index = hier_index, columns = ['A', 'B'])
```

```python
df
```

```python
df.loc['G1']
```

```python
df.index.names = ['Grupo', 'Numero']
```

```python
df
```

```python
df.xs(1, level = 'Numero')
#o xs (cross section) pode selecionar de maneira melhor que o loc
#pois com ele conseguimos sselecionar tambem niveis internos
```
