```python
import pandas as pd
import numpy as np
```

```python
d = {'A': [1,2,np.nan], 'B': [5,np.nan,np.nan], 'C': [1,2,3]}
#IMPORTANTE: dicionarios usam chaves
```

```python
d
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df.dropna() #exclui todas as linhas que tem valores invalidos
```

```python
df.dropna(thresh=2)
#com o thresh ele so exclui onde tiver o numero indicado de valores invalidos
```

```python
df #em qualquer ferramenta pandas so e feita alteracao permanente com a aplicacao do inplace
```

```python
x =  np.random.randint(1,35)
df.fillna(value = x)
#substitui os valoes invalidos pelo valor desejado
```

```python
df.fillna(method = 'ffill')
#esse metodo substitui replivcando o elemento anterior (forward fill)
```
